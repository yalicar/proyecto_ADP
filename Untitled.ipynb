```python
import numpy as np
y = np.arange(1, 3)
x = np.array([2, 4, 6, 8])
x=np.array(x).reshape(-1,1)
y=np.array(y).reshape(-1,1)
x, y
```

```python
def prediccion(x,y):
    x=np.array(x)
    x=np.hstack([x,np.ones_like(x)])
    pred=x.dot(y)
    return pred
```

```python
resp=prediccion(x,y)
```
