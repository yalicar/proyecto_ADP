## 2. Al usar Pandas loc y iloc nos sirven para?



- Se utilizan para seleccionar filas y columnas de un Dataframe

```python
#Codigo de ejemplo
import pandas as pd

data = {'Country': ['Belgium',  'India',  'Brazil'],'Capital': ['Brussels',  'New Delhi',  'Brasilia'],'Population': [11190846, 1303171035, 207847528]}
df = pd.DataFrame(data,columns=['Country',  'Capital',  'Population'])
df.head() # visualizar las primeras filas del dataframe
df.iloc[:1] # Seleecion del primer registro
df.iloc[:,0:1] # seleccion de la primera columan y todas las filas
```

## 4. Para calcular la transpuesta de un dataset, tanto en NumPy como en Pandas podemos usar?

- Para numpy se puede usar np.transpose()
- Para pandas se puede usar df.T

```python
import numpy as np
import pandas as pd
a=[1,2,3,4,]
a=np.reshape(np.array(a),(2,2))
a
```

```python
g=a.transpose() # Transpuesta usando Numpy
g
```

```python
h=pd.DataFrame(g)
h
```

```python
h_t=h.T # Transpuesa usando pandas
h_t
```
