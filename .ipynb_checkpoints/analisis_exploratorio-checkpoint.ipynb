# Para cada variable independiente x

## Calcular el coeficiente de correlaci√≥n entre x y y.

```python
sns.heatmap(data.corr(), annot = True)
plt.show()
```

```python
#Calculo de los valores de correlacion para cada par de variables.
x={}
for i in range(0,5):
    for j in range(0,5):
        name=str(data.columns[i])+ "/"+ str(data.columns[j])
        x.update({name:data.iloc[:,i].corr(data.iloc[:,j])})
df_x = pd.DataFrame([[key, x[key]] for key in x.keys()], columns=['Name', 'Correlacion']) # conversion de diccionario a dataframe
df_x
```

## Graficar x vs y(scatterplot) usando matplotlib.

```python
k=0
for i in range(0,5):
    for j in range(0,5):
        plt.scatter(data.iloc[:,j],data.iloc[:,i])
        plt.xlabel(data.columns[j])
        plt.ylabel(data.columns[i])
        plt.title(f" correlacion {df_x.iloc[k,0]} : {x[df_x.iloc[k,0]]}")
        k=k+1
        plt.show()
```

## Basado en la gr√°fica y el coeficiente de correlaci√≥n de cada par x,y elegir las 2 variables con m√°s potencial predictivo es decir las 2 variables que presentan mayor correlaci√≥n entre dicha variable y la variable dependiente.

```python
df_x.sort_values(by=['Correlacion'], inplace=True,ascending=False)
df_x[df_x['Correlacion'] < 0.99].head(10)
# Las variables que poseen mayor correlacion con la variable dependiente son "OverallQual" y "stFlrSF".
```
